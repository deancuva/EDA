# **3. Load the data using Pandas. Clean them up for EDA. Do this in a notebook with comments or markdown chunks explaining your choices.**

```python
df = pd.read_csv(output_filename) #load data with my selected variables
print(df.info()) #print data to ensure that the correct variables are selected and view variable information. We see that the variables match with our selected data.
print(df.head()) #print first few rows of data
```

We see that there are many instances of missing data, especially within the mapres10, rincome, and hrs2 variables. To improve data quality, we want to drop rows with missing data.

```python
df = df.dropna(subset=['rincome', 'mapres10', 'hrs2']) #we want to drop rows with missing data, specifically in the mapres10, rincome, and hrs2 variables. If we drop all rows with missing data from each variable, the dataset may be empty (as there are many missing values).
print(df.head()) #print the first 5 rows to examine changes
```

Next, we want to convert all numeric variables to floats, and all categorical variables to categories. This will allow us to further analyze the data.

```python
df['papres10'] = pd.to_numeric(df['papres10'], errors='coerce') #Convert numeric values
df['mapres10'] = pd.to_numeric(df['mapres10'], errors='coerce')
df['hrs2'] = pd.to_numeric(df['hrs2'], errors='coerce')
df['agewed'] = pd.to_numeric(df['agewed'], errors='coerce')

df['degree'] = df['degree'].astype('category') #Convert categorical variables
df['rincome'] = df['rincome'].astype('category')
df['marital'] = df['marital'].astype('category')
df['padeg'] = df['padeg'].astype('category')
df['madeg'] = df['madeg'].astype('category')

print(df.dtypes) #Examine the datatypes to ensure accuracy
```

```python
df['degree'] = df['degree'].str.lower() #convert all strings to lowercase to ensure standardization
df['rincome'] = df['rincome'].str.strip().str.lower()
print(df.head())
```

```python
#Produce numeric summaries

#Print summary statistics
summary_stats = df.describe()
print(summary_stats)

#Correlation between papres10 and mapres10
correlation = df['papres10'].corr(df['mapres10'])
print(f"Correlation between papres16 and mapres10: {correlation:.2f}")

#Percentage of people in each marriage status group
marital_percentage = df['marital'].value_counts(normalize=True) * 100
print("\nPercentage of People in Each Marital Status:")
print(marital_percentage)

#Descriptive stats for papres 10 grouped by level of degree held
degree_grouped = df.groupby('degree')['papres10'].describe()
print("\nDescriptive statistics for 'papres10' by 'degree':")
print(degree_grouped)

#Income by prestige of occupation held by respondent's Mother
income_occupation = df.groupby('rincome')['mapres10'].describe()
print("\nDescriptive statistics for 'mapres10' by 'rincome':")
print(income_occupation)

#Stats for marriage status grouped by agewed
marital_agewed = df.groupby('marital')['agewed'].describe()
print("\nDescritive statistics for Agewed by Marital Status")
print(marital_agewed)
```

# **5. Describe your findings in 1-2 pages.**

After using our posed questions to create graphs that visually delivered answers, we found some interesting results. Firstly, prior to creating graphs, we expected that there would be a positive correlation between parental education level and marital age. We hypothesized this by, first, assuming that children of those who received a college degree were more likely to receive one themselves. Then, because the vast majority of our fellow classmates are not married, we intuited that those who have parents educated to a higher degree would be more likely to marry later in life. However, we found this to not be the case in either scenario. As the education level of a respondent’s mother and father increased, the more likely they were to become married early in life. We were initially shocked by this finding, but have since rationalized that this might be because those who get married often need resources to do so and to support them following their union. Thus, those who get married at a younger age may have more affluent parents, and there is a very positive correlation between education-level and wealth. Albeit, the variation between the three levels of education is marginally greater than a single year, so not a significant disparity.

In terms of parental job prestige’s relationship with respondent income, we again found an insignificant correlation between the two, especially at the moderate to low levels of prestige. There were peaks that appeared seemingly at random above and below the median income level, meaning that there was no significant sway caused by a parent’s job prestige on the income of their child. Additionally, it is important to note the disjointed nature of job prestige and income, as a job can be prestigious without being very high-paying and vice versa–for instance, academia and certain governmental institutions. Therefore, it would be interesting in future research to delve into whether parental job prestige had any impact on the respondent’s job prestige.

For the relationship between parents’ level of education and hours worked, there appeared to be minute differences in the median number of hours worked and trends proved to be essentially identical between mothers and fathers. The hours stayed relatively constant at forty hours a week, but it should obviously be noted that this is considered the national standard, and thus deviations are rare and may be diluted in the data. The aspect of note is the weekly average range for the different education levels. If a respondent’s father has a bachelor’s degree, the range is greatest, which can probably be attributed to the child likely being afforded more opportunities, and potentially working in more demanding industries: such as professional services, finance, or technology. Another interesting component of the boxplot is that for father;s with a graduate level education, there is virtually no consistent logging of more than forty hours a week for respondents. This is probably due to the fact that those with graduate level degrees have demonstrable expertise and may be wealthier due to the compensation that often comes with that. Thus, their children don’t need to go into such demanding industries. Furthermore, the most outliers in hours worked came from respondents with parents with only a highschool education. Because their parents didn’t receive a college degree, they are much less likely as well, and thus may have entered industries that, while less prestigious, are still very demanding, like the trades. education as well.

Through this analysis, many of our initial hypotheses were disproven and from that sprang new questions that we would have explored further given a larger project scope. Ultimately, we decided that in the context of the limited scope of our project, we conclude that there is not a significant disparity in the lifestyles of respondents based on certain attributes of their parents. This was unexpected, but of course, there are several caveats and limitations that must be addressed. Firstly, there are many immeasurable factors that play into the eventual lifestyles of an individual, regardless of their parents, which were obviously not taken into account. Moreover, we only examined a handful of variables and their relationships with another limited set of variables. Additionally, we operated under certain assumptions that already factored in lifestyle disparities, such as that well-educated parents earned more than parents without as much as an education. Finally, there is obviously great nuance to all of these variables, which was not addressed in our research. For instance, the quality of a college degree certainly factors into eventual success at a macro scale, as in a degree from a for-profit institution is not equivalent to an Ivy League degree. Thus, this topic warrants further exploration.
